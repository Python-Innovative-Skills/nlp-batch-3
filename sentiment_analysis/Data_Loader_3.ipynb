```python
import numpy as np
import pandas as pd
import codecs
```

```python
INPUT_FILE = "ecommerce_dataset.txt"
fin = codecs.open(INPUT_FILE, "r", encoding='utf-16')
for line in fin:
  
    label, sent = line.strip().split("\t") #Stripping the dataset based on tab. That is stripping label from sentence
    print("Label: ", label)
    print("Sentence: ",sent)
```

```python
with open(INPUT_FILE, "r", encoding='utf-16') as fin:
    for line in fin:
        _, sent = line.strip().split("\t")  # Stripping the dataset based on tab
        print("Label: ", _)
        print("Sentence: ", sent)
```
